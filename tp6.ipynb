```python
from tkinter import *
from tkinter import filedialog
from tkinter import Tk, Button, Canvas
from PIL import Image,ImageTk
import PIL
import cv2
import numpy as np
from matplotlib import pyplot as plt
from sklearn import tree
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
from sklearn.neural_network import MLPClassifier
import timeit


#les images d apprentissage
y=[]
# l image a laquel en vas faire du teste
X=[]

decision=''
multi=''
cl=''

for i in range(1,63):
    img = cv2.imread("C:/Users/DELL/Desktop/TAD/images/apprentissage/".__add__(i.__str__()).__add__(".png"),0)
    th2 = cv2.adaptiveThreshold(img,1,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
    Xi=th2.flatten()
    y.append(Xi)
    
d=["A","A","A","A","A","B","B","B","C","C","D","D","E","E","E","F","F","G","G","H","H","I","I","I","I","I","J","J","K","K","L","L","M","M","N","N","O","O","O","P","P","Q","Q","R","R","S","S","T","T","U","U","U","V","V","W","W","X","X","Y","Y","Z","Z"]
def DecisionTreeClassifier(d):
    global decision
    decision = tree.DecisionTreeClassifier(ccp_alpha=0.005,criterion='entropy',splitter='random',random_state=5)
    decision.fit(y,d)
    return decision

#l apprentissage avec multinomial  
def multinomial(d):
    global multi
    multi = MultinomialNB()
    multi.fit(y,d)
    return multi

def multicouche(d):
    global cl
    cl=MLPClassifier(max_iter=10000)
    cl.fit(y,d)
    return cl



DecisionTreeClassifier(d)
multinomial(d)
multicouche(d)



def temp_Apprentissage():     
    temp_dtc = timeit.Timer('DecisionTreeClassifier(d)','from __main__ import DecisionTreeClassifier, d')
    temp_multi = timeit.Timer('multinomial(d)','from __main__ import multinomial, d')
    temp_multi_couche = timeit.Timer('multicouche(d)','from __main__ import multicouche, d')
    return temp_dtc,temp_multi,temp_multi_couche


a,b,c = temp_Apprentissage() 
print(a.timeit(1))
print(b.timeit(1))
print(c.timeit(1))



te=["A","A","B","B","C","C","D","D","E","E","F","F","Z","G","H","H","I","I","J","J","Y","K","X","L","M","M","N","N","O","O","P","P","Q","W","V","R","S","S","U","T","U"]
s=0
s_multi=0
s_multi_couche=0
for i in range(1,42):
    img = cv2.imread("C:/Users/DELL/Desktop/TAD/images/test/".__add__(i.__str__()).__add__(".png"),0)
    th2 = cv2.adaptiveThreshold(img,1,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
    Ki=th2.flatten()
    
    if(te[i-1]==decision.predict([Ki])):
        s=s+1
    if(te[i-1]==multi.predict([Ki])):
        s_multi=s_multi+1
    if(te[i-1]==cl.predict([Ki])):
        s_multi_couche=s_multi_couche+1
        
taux_app_decision = s/41*100
taux_app_multinomial = s_multi/41*100
taux_app_multicouche = s_multi_couche/41*100


print(taux_app_decision)
print(taux_app_multinomial)
print(taux_app_multicouche)



root = Tk() 


partition11 = LabelFrame(root,text="comparaison des TAAD")
partition11.pack(side=TOP,expand="yes",padx=10,pady=10)
name1 = Label(partition11,text = "comparaison des trois methodes")
name1.pack(padx=10,pady=10)
p = PanedWindow(partition11, orient=VERTICAL)
p.pack(side=LEFT, pady=20, padx=20)
p.add(Label(p, text='classifier', background='white', anchor=CENTER))
p.add(Label(p, text='MLP CLASSIFEIER', background='white', anchor=CENTER) )
p.add(Label(p, text='MULTINOMIAL', background='white', anchor=CENTER) )
p.add(Label(p, text='DecisionTreeClassifier', background='white', anchor=CENTER) )
p.pack()
p2 = PanedWindow(partition11, orient=VERTICAL)
p2.pack(side=LEFT, pady=20, padx=20)
p2.add(Label(p2, text='temps d execution ', background='white', anchor=CENTER))
p2.add(Label(p2, text=a.timeit(1), anchor=CENTER) )
p2.add(Label(p2, text=b.timeit(1), anchor=CENTER) )
p2.add(Label(p2, text=c.timeit(1), anchor=CENTER) )
p2.pack()
p3 = PanedWindow(partition11, orient=VERTICAL)
p3.pack(side=LEFT, pady=20, padx=20)
p3.add(Label(p3, text='Taux apprentissage', background='white', anchor=CENTER))
p3.add(Label(p3, text=taux_app_multicouche, anchor=CENTER) )
p3.add(Label(p3, text=taux_app_multinomial, anchor=CENTER) )
p3.add(Label(p3, text=taux_app_decision, anchor=CENTER) )
p3.pack()

root.title("apprentissage des alphabets")
root.geometry("500x250")
root.mainloop()

```
